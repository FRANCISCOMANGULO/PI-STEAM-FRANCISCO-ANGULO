```python

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Convertir la columna "review" a cadenas de texto
df_user_reviews_retocado['review'] = df_user_reviews_retocado['review'].astype(str)

# Crear una instancia del analizador de sentimientos de VaderSentiment
analyzer = SentimentIntensityAnalyzer()

# Crear una nueva columna "sentimiento"
df_user_reviews_retocado['Sentimiento'] = df_user_reviews_retocado['review'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# Imprimir el DataFrame con la nueva columna "sentimiento"
print(df_user_reviews_retocado.head(5))
```

```python


def etiquetar_sentimiento(texto):
    # Crear una instancia del analizador de sentimientos de VaderSentiment
    sia = SentimentIntensityAnalyzer()

    # Analizar el sentimiento del texto
    sentiment = sia.polarity_scores(texto)
    compound_score = sentiment['compound']

    # Asignar una etiqueta de sentimiento de acuerdo con el puntaje compuesto
    if pd.isnull(texto) or texto == '':
        return 1  # Return neutral if it is empty or NaN
    elif compound_score >= 0.05:
        return 2  # Buena puntuaci칩n
    elif compound_score <= -0.05:
        return 0  # Malas puntuaciones
    else:
        return 1  # Neutral puntuaciones
```

```python
# Crear una nueva columna "etiqueta"
df_user_reviews_retocado['etiqueta'] = df_user_reviews_retocado['review'].apply(etiquetar_sentimiento)
```
