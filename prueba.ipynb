```python
def UserForGenre(genero):
    # Leer los DataFrames necesarios
    df = pd.read_parquet(['User_reviews_reducido_32.parquet', 'Output_steam_games_reducido_32.parquet', 'User_items_reducido_32.parquet', 'Top 3 Desarrolladores.parquet', 'User_data.parquet'])
    
    # Combinar los DataFrames necesarios
    df_merged = pd.merge(df[1], df[2], on='Item_Id')
    df_nuevo = df_merged.drop(columns=['Item_Id','App_name','Price','Developer','Items_Count','Item_Name'])
    df_f1 = df[1][["Item_Id", "Price","Developer","Release_year"]]
    df_limpio = pd.merge(df[1], df[0], on='Item_Id')[['User_Id', 'Item_Id','Developer','Release_year','Sentiment Analysis']]
    
    # Realizar la operación de agrupación
    if not genero in df_nuevo.columns:
        return f"El género {genero} no existe en la base de datos."
    
    df_genre = df_nuevo[df_nuevo[genero] == 1]
    usur_mas_horas = df_genre.groupby("User_Id")["Playtime_Forever"].sum().idxmax()
    filtro_usur = df_genre[df_genre["User_Id"] == usur_mas_horas]
    horas_jugXaño = filtro_usur.groupby("Release_year")["Playtime_Forever"].sum().to_dict()
    
    # Liberar la memoria utilizada por los DataFrames que ya no se necesitan
    del df, df_merged, df_nuevo, df_f1, df_limpio, df_genre, filtro_usur, horas_jugXaño
    gc.collect()
    
    # Devolver los resultados
    return {"Usuario con más horas jugadas": usur_mas_horas, "Horas jugadas por año": horas_jugXaño}
```

```python
def UserForGenre(genero):
    if not genero in df_nuevo.columns:
        return f"El género {genero} no existe en la base de datos."
    
    df_genre = df_nuevo[df_nuevo[genero] == 1]

    usur_mas_horas = df_genre.groupby("User_Id")["Playtime_Forever"].sum().idxmax()

    filtro_usur = df_genre[df_genre["User_Id"] == usur_mas_horas]

    horas_jugXaño = filtro_usur.groupby("Release_year")["Playtime_Forever"].sum()

    registro = horas_jugXaño.to_dict()

    Horas_por_año = {}
    for clave, valor in registro.items():
        clave_formateada = f'Año: {int(clave)}'
        valor_formateado = int(valor)
        Horas_por_año[clave_formateada] = valor_formateado
    del df_genre, filtro_usur, horas_jugXaño
    gc.collect()
    return {"Usuario con más horas jugadas": usur_mas_horas, "Horas jugadas por año": Horas_por_año}
```

```python
import pandas as pd
import pyarrow.parquet as pq
import gc

def top_desarrolladores_recomendados(year):
    # Leer los archivos Parquet necesarios
    df8 = pq.read_table('Output_steam_games_reducido_32.parquet').to_pandas()
    df9 = pq.read_table('User_items_reducido_32.parquet').to_pandas()
    merged_df = pq.read_table('Top 3 Desarrolladores.parquet').to_pandas()
    
    # Realizar las operaciones necesarias con los DataFrames
    merged_df10 = pd.merge(df8, df9, on='Item_Id')
    df_year = merged_df[merged_df['year'] == year]
    df_count = df_year[df_year['Recommend'] == True][df_year['Sentiment Analysis'] == 2].groupby('Developer')['App_name'].count().reset_index()
    top_desarrolladores = df_count.sort_values('App_name', ascending=False).head(3)['Developer'].tolist()

    # Eliminar los DataFrames que ya no se necesitan
    del df8, df9, merged_df, merged_df10, df_year, df_count
    gc.collect()

    # Devolver el top 3 de desarrolladores
    return top_desarrolladores
```
