```python
# Leer el archivo JSON en un DataFrame
df = pd.read_json(r'C:\Users\baby_\Downloads\PI MLOps - STEAM-20231023T194033Z-001\PI MLOps - STEAM\steam_games.json\output_steam_games.json', lines=True)

df.columns = df.columns.str.capitalize()

# Imprimir el DataFrame
print(df.tail(5))
```

```python
# Eliminar todas las filas que contengan valores nulos en la columna "id"
df = df.dropna(subset=['Id'])

# Imprimir el DataFrame resultante
print(df.head(5))
```

```python
# Seleccionar las filas con valores nulos en la columna "genres"
null_genres = df[df['Price'].isnull()]

# Imprimir las filas seleccionadas
print(null_genres)
```

```python
import pandas as pd

# Eliminar las columnas "Title" y "Url" del DataFrame
df = df.drop(['Title', 'Url','Reviews_url','Early_access','Tags'], axis=1)

# Imprimir el DataFrame resultante
print(df)
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df['Genres'] = df['Genres'].fillna('[]')  # Rellenar los valores faltantes con una lista vacía
df['Genres'] = df['Genres'].apply(lambda x: ', '.join(x))  # Convertir la lista de géneros a una cadena separada por comas
```

```python
dummy_genres = df['Genres'].str.get_dummies(', ')
```

```python
df_steam_games = pd.concat([df, dummy_genres], axis=1)
```

```python
df_steam_games = df_steam_games.drop(columns=['[', ']'])

# Imprimir el DataFrame resultante
df_steam_games
```

```python

# Extraer conjuntos de 4 números de la columna "Release_date"
df_steam_games['Release_year'] = df_steam_games['Release_date'].str.extract(r'(\d{4})')

# Imprimir el DataFrame resultante
print(df_steam_games)
```
