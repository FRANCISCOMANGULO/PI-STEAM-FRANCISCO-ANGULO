### EDA Top Developer por user

```python
review_limpio = reviews_steams.drop(columns=["año","Release_year"])
```

```python
dev_usuarios = review_limpio.groupby("Developer")["User_Id"].count().sort_values(ascending=False).reset_index()
```

```python
top_5 = dev_usuarios.head(5)
```

```python
# Crea un gráfico de barras utilizando Seaborn
sns.set(style="whitegrid")
ax = sns.barplot(x="Developer", y="User_Id", data=top_5, palette="muted")

# Agrega etiquetas y título
ax.set(xlabel='Developer', ylabel='User_Id', title='Top 5 Developers por User')

# Rota las etiquetas del eje x
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Muestra el gráfico
plt.show()
```

### EDA Distribucion de analisis de sentimientos por usuario

```python
review_limpio.head(5)
```

```python
# Contar el número de sentimientos de análisis positivos, negativos y neutrales
positivas = len(review_limpio[review_limpio["Sentiment Analysis"] == 2])
neutrales = len(review_limpio[review_limpio["Sentiment Analysis"] == 1])
negativos = len(review_limpio[review_limpio["Sentiment Analysis"] == 0])

# Crear un gráfico de pastel utilizando Matplotlib
plt.figure(figsize=(6, 6))
colors = sns.color_palette("pastel")
plt.pie([positivas, negativos, neutrales], labels=["Positivas", "Negativas", "Neutral"], autopct='%1.1f%%', startangle=90, colors=colors)

# Agregar título
plt.title("Distribucion por sentimiento de análisis")

# Mostrar el gráfico
plt.show()
```

### EDA Top 5 Juegos mas recomendados

### EDA Genero mas populares por los users

```python
items_steam.info(max_cols=None)
```

```python
genero_jug = items_steam[['Accounting','Action', 'Adventure', 'Animation &amp; Modeling','Casual','Design &amp; Illustration', 'Early Access','Education','Free to Play', 'Indie', 'Massively Multiplayer','Photo Editing', 'RPG', 'Racing', 'Simulation','Software Training', 'Sports', 'Strategy','Utilities', 'Video Production', 'Web Publishing']]
```

```python

# Ordenar los géneros de mayor a menor
pastel_gen_ordenado = pastel_gen.sort_values(ascending=True)

# Crear un gráfico de barras horizontales utilizando Matplotlib
plt.barh(pastel_gen_ordenado.index, pastel_gen_ordenado)

# Agregar título y etiqueta del eje y
plt.title('Géneros más populares')
plt.ylabel('Género')

# Mostrar el gráfico
plt.show()
```

### EDA Juegos con mas tiempo jugado

```python
juego_mas = items_steam[[ "App_name","Playtime_Forever"]] 
```

```python
juego_mas = juego_mas.groupby("App_name")["Playtime_Forever"].sum().reset_index()
```

```python
topj = juego_mas.sort_values("Playtime_Forever",ascending=False)
```

```python
juego55 = topj.head(5)    
```

```python
juego55
```

```python
import matplotlib.pyplot as plt

# Convertir el tiempo de juego a horas
juego55['Playtime_Forever'] = juego55['Playtime_Forever'] / 3600

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Crear el gráfico de barras
ax.bar(juego55['App_name'], juego55['Playtime_Forever'])

# Rotar las etiquetas del eje x para que sean legibles
plt.xticks(rotation=90)

# Agregar título y etiquetas de los ejes
ax.set_title('Top 5 juegos con más tiempo jugado')
ax.set_xlabel('Juego')
ax.set_ylabel('Tiempo jugado (horas)')

# Mostrar el gráfico
plt.show()
```
