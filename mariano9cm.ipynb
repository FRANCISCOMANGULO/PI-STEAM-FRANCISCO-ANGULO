```python
import pandas as pd

def UserForGenre(genero):
    # Lee el archivo CSV en trozos de 5000 filas
    chunks = pd.read_csv('User_for_genre.csv', chunksize=5000)

    # Itera sobre los trozos y procesa los datos
    usur_mas_horas = None
    for chunk in chunks:
        if not genero in chunk.columns:
            return f"El género {genero} no existe en la base de datos."
        
        df_genre = chunk[chunk[genero] == 1]

        if df_genre.empty:
            continue

        usur_mas_horas_chunk = df_genre.groupby("User_Id")["Playtime_Forever"].sum().idxmax()

        if usur_mas_horas is None or df_genre[df_genre["User_Id"] == usur_mas_horas_chunk]["Playtime_Forever"].sum() > df_genre[df_genre["User_Id"] == usur_mas_horas]["Playtime_Forever"].sum():
            usur_mas_horas = usur_mas_horas_chunk

    if usur_mas_horas is None:
        return f"No se encontraron datos para el género {genero}."

    filtro_usur = df_genre[df_genre["User_Id"] == usur_mas_horas]

    horas_jugXaño = filtro_usur.groupby("Release_year")["Playtime_Forever"].sum()

    registro = horas_jugXaño.to_dict()

    Horas_por_año = {}
    for clave, valor in registro.items():
        clave_formateada = f'Año: {int(clave)}'
        valor_formateado = int(valor)
        Horas_por_año[clave_formateada] = valor_formateado
    del df_genre, filtro_usur, horas_jugXaño
    gc.collect()
    return {"Usuario con más horas jugadas": usur_mas_horas, "Horas jugadas por año": Horas_por_año}
```
