```python
import pandas as pd    # Para manejo de datos
import numpy as np    # Para operaciones matemáticas
import matplotlib.pyplot as plt     # Para gráficos
import seaborn as sns   # Para gráficos más bonitos
import os   # Para manejo de archivos
import datetime as dt   # Para manejo de fechas
import json   # Para manejo de archivos json
import ast    # Para manejo de archivos json
import re    # Para manejo de expresiones regulares
import jsonlines   # Para manejo de archivos jsonlines
```

```python

filas = list()

with open(r"C:\Users\baby_\Downloads\PI MLOps - STEAM-20231023T194033Z-001\PI MLOps - STEAM\users_items.json\australian_users_items.json", "r",encoding="utf-8") as archivo:
    for linea in archivo:
        filas.append(ast.literal_eval(linea))
```

```python
df2 = pd.DataFrame(filas)
```

```python
df2['items'].isnull().sum()
```

```python
df2 = df2.drop(columns=['user_url','steam_id'])
```

```python


# Hacer "explode" a la columna "items"
df2 = df2.explode('items')

# Imprimir el DataFrame resultante
print(df2)
```

```python
df2 = df2.reset_index(drop=True)
```

```python


# Utilizar la función json_normalize() para convertir los objetos JSON en un DataFrame plano
df_items = pd.json_normalize(df2['items'])

# Concatenar el DataFrame resultante con el DataFrame original
df2 = pd.concat([df2, df_items], axis=1)
```

```python
df2
```

```python


# Eliminar la columna "playtime_2weeks"
df2 = df2.drop(columns=['playtime_2weeks','items'])
df2
```

```python
df2 = df2.rename(columns=lambda x: x.title())
df2
```

```python
df2 = df2.dropna(subset=["Item_Id"])
```
